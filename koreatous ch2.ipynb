```python
uslife = df1['왜 미국 생활 만족/불만족하는 이유?']
uslife =uslife.dropna()
uslife
```

```python
# Python 유저가 한글 자연어 처리를 할 때 필수 모듈인 KoNLPy를 소개합니다.
# http://pinkwink.kr/1025
import nltk
from konlpy.corpus import kolaw
```

```python
# 한국어 중 25개의 가장 많이 출현하는 단어수
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize

stop_words = set(stopwords.words('english'))
stop_words.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','것','수','및','..?','&','/','있는','다','-']) # remove it if you need punctuation 

tokens_uslife = wordpunct_tokenize(uslife_txt.lower())
tokens_uslife = [each_word for each_word in tokens_uslife if each_word not in stop_words]

ko = nltk.Text(tokens_uslife, name='미국 생활 만족도')
usfreq = ko.vocab().most_common(25)
topwords = usfreq[0:15]
topwords
```

```python
# 단어 빈도수 표시
dicwords = dict(topwords)

twords_inx = dicwords.keys()
y_pos = np.arange(len(twords_inx))
twords_cnt = dicwords.values()

plt.bar(y_pos, twords_cnt, align='center', alpha=1)
plt.ylim(1,15)
plt.xticks(y_pos,twords_inx, fontproperties=fontprop, rotation=45)
plt.title('미국 생활 만족도 단어 빈도', fontproperties=fontprop)
plt.ylabel('빈도 수', fontproperties=fontprop)
plt.show()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize

uslife_txt = uslife_txt.replace('balance가','balance')
uslife_txt = uslife_txt.replace('기회가','기회')
uslife_txt = uslife_txt.replace('만족도가','만족도')
uslife_txt = uslife_txt.replace('가족과','가족')
uslife_txt = uslife_txt.replace('한국에','한국')
uslife_txt = uslife_txt.replace('한국서는','한국')
uslife_txt = uslife_txt.replace('한국이','한국')
uslife_txt = uslife_txt.replace('시간과','시간')
uslife_txt = uslife_txt.replace('가족혹은','가족')
uslife_txt = uslife_txt.replace('가족들의','가족')
uslife_txt = uslife_txt.replace('가족이','가족')
uslife_txt = uslife_txt.replace('저녁과','저녘')
uslife_txt = uslife_txt.replace('저녁이','저녘')

token_uslife = wordpunct_tokenize(uslife_txt.lower())

stop_words = set(stopwords.words('english'))
stop_words.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','것','수','및','..?','&','-','/','1)','2)','3)','),','...','있는']) # remove it if you need punctuation 

wordlist = [each_word for each_word in token_uslife if each_word not in stop_words]

wordfreq = []
for w in wordlist:
    wordfreq.append(wordlist.count(w))
 
# 단어 빈도수를 사전으로 변경
dicwords = dict(zip(wordlist,wordfreq))
dicwords
```
