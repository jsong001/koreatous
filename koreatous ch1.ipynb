```python
# https://github.com/amueller/word_cloud/blob/master/wordcloud/wordcloud.py
# font_path=None, width=400, height=200, margin=2,
#                 ranks_only=None, prefer_horizontal=.9, mask=None, scale=1,
#                 color_func=None, max_words=200, min_font_size=4,
#                 stopwords=None, random_state=None, background_color='black',
#                 max_font_size=None, font_step=1, mode="RGB",
#                 relative_scaling=.5, regexp=None, collocations=True,
#                 colormap=None, normalize_plurals=True, contour_width=0,
#                 contour_color='black'
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
krlife = df1['어떤 면에서 한국 생활 만족/불만족 하시나요?']
krlife = krlife.dropna()
krlife
```

```python
krlifeinx = krlife.index

krlife_tmp = []
for n in krlifeinx: 
     krlife_tmp.append(krlife[n])

krlife_txt = ' '.join(krlife_tmp)
krlife_txt 
```

```python
# 간편하게 사용하면서도 꽤 이쁜 기능도 많은 워드 클라우드 Word Cloud 모듈 (for Python)
# http://pinkwink.kr/1029
krmap_coloring = np.array(Image.open("Korea_Map.jpg"))
stopwords = set(STOPWORDS)
stopwords.add("그러나")
                          

wcloud = WordCloud(font_path='C:\\WINDOWS\\Fonts\\NanumMyeongjo.ttf',
                  relative_scaling = 0.2,
                  background_color='white',mask=krmap_coloring,stopwords=stopwords,
                  max_font_size=40).generate(krlife_txt)

plt.figure(figsize=(12,12))
plt.imshow(wcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Python 유저가 한글 자연어 처리를 할 때 필수 모듈인 KoNLPy를 소개합니다.
# http://pinkwink.kr/1025
import nltk
from konlpy.corpus import kolaw
```

```python
# 한국어 중 25개의 가장 많이 출현하는 단어수
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize

stop_words = set(stopwords.words('english'))
stop_words.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','것','수','및','..?','&','/','있는','다','등']) # remove it if you need punctuation 

tokens_krlife = wordpunct_tokenize(krlife_txt)
tokens_krlife = [each_word for each_word in tokens_krlife if each_word not in stop_words]

ko = nltk.Text(tokens_krlife, name='한국 생활 만족도')
krfreq = ko.vocab().most_common(25)
```

```python
topwords = krfreq[0:11]
topwords
```

```python
# 단어 빈도수 표시
dicwords = dict(topwords)

twords_inx = dicwords.keys()
y_pos = np.arange(len(twords_inx))
twords_cnt = dicwords.values()

plt.bar(y_pos, twords_cnt, align='center', alpha=1)
plt.ylim(1,10)
plt.xticks(y_pos,twords_inx, fontproperties=fontprop, rotation=45)
plt.title('한국 생활 만족도 단어 빈도', fontproperties=fontprop)
plt.ylabel('빈도 수', fontproperties=fontprop)
plt.show()
```
